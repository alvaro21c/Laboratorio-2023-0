```python
# Visualizacion de datos 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y)
plt.show()
```

```python
# editar contenido de nuestro grafico
plt.plot(x,y)
plt.xlabel('eje x') # para ponerle nombre al eje horizontal
plt.ylabel('eje y') # para ponerle nombre al eje vertical
plt.title('Mi primer grafico') # para ponerle titulo
```

```python
x1=[11,12,13,14,15]
plt.plot(x,y)
plt.plot(x,x1)
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.title('Segundo grafico')
plt.legend(['(x,y)','(x,x1)'],loc='lower right')
plt.grid ()
plt.show()
```

```python
plt.plot(x,y)

# agregar rango para eje x e y
plt.axis([0,6,0,20]) # como se modifica el rango. Los dos primeros son para x y los dos ultimos y
plt.show()
```

```python
# tipos de linea para el grafico
plt.figure(figsize=(10,2))
plt.subplot(131)
plt.plot(x,y,"-.")

plt.subplot(132)
plt.plot(x,y,'o')

plt.subplot(133)
plt.plot(x,y,'.')

plt.show()
```

```python
# colores del grafico
plt.figure(figsize=(10,2))
plt.subplot(131)
plt.plot(x,y,"r-.")

plt.subplot(132)
plt.plot(x,y,'bo')

plt.subplot(133)
plt.plot(x,y,'k.')

plt.show()
```

```python
# graficos con operaciones
a=np.arange(0,5,0.2)
plt.plot(a,a,'r-.',a,a**2,'bo',a,a**3,'k.')

# poner leyenda queda pendiente 
```

```python
# grafico de barras
c=['Melgar','Cienciano','Cristal','Alianza','Universitario']
plt.bar(c,y)
```

```python
# grafico de barras 2

clases=['Economia','Finanzas','Gestion','Derecho']
h=[60,50,90,85]
m=[75,55,88,49]
adultosmy=[75,55,88,49]

a=len(clases)
x=np.arange(a)

ancho= 0.3

plt.bar(x-ancho,h,width=ancho,label='h') # nuestro primer grafico
plt.bar(x,m,width=ancho,label='m') # 2do graf
plt.bar(x+ancho,adultosmy,width=ancho,label='my')
# segundo grafico 
plt.xticks(x,clases)
plt.legend()
plt.show()

```
