```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%plt.inline # que los graficos se encuentren dentro del cuaderno
```

```python
propinas=sns.load_dataset('tips') # este es el codigo que carga los datos que existen en seaborn
propinas.head() 
# planets, titanic, iris, flights
```

```python
sns.relplot(data=propinas,x='total_bill',y='tip')
```

```python
sns.relplot(data=propinas,x='total_bill',y='tip',hue='sex')
```

```python
sns.relplot(data=propinas,x='total_bill',y='tip',hue='day')
```

```python
sns.relplot(data=propinas,x='total_bill',y='tip',hue='day',col='sex')
```

```python
sns.relplot(data=propinas,x='total_bill',y='tip',hue='size')
plt.show()
```

```python
# Graficos de distribucion
sns.displot(data=propinas,x='tip')
```

```python
# curva de densidad
sns.displot(data=propinas,x='tip',kind='kde')
```

```python
sns.displot(data=propinas,x='tip',col='time')
```

```python
# Grafico de cajas
sns.catplot(data=propinas,x='day',y='total_bill',kind='box')
```

```python
sns.catplot(data=propinas,x='day',y='total_bill',kind='bar')
```

```python
sns.catplot(data=propinas,x='day',y='total_bill',kind='bar',ci=False) # mismo graf sin error bars
```

```python
sns.catplot(data=propinas,x='day',y='total_bill',kind='bar',hue='sex')
```

```python
sns.jointplot(data=propinas,x='total_bill',y='tip',kind='scatter')
```
