```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
x = np.arange(11)
y = np.arange(0,1001,100)
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
# o,*,.,x,X,+,P,s,D,d,p,H,h,v,>,<,1,2,3,4,_

l=['o','*','x','+','s','p','1','2','3','4','>','<']

for m in l:
    print(m)
    plt.plot(x,m)
    plt.show()
```

```python
linea = ['-',':','--','-.']

for l in linea:
    print (l)
    l='o'+l+'r' # estoy concatenando. En el primer caso tendria o-r
    plt.plot(y,l)
    plt.show()
```

```python
colores=['r','g','b','c','m','y','k','w']

for c in colores:
    print (c)
    c='o:'+c
    plt.plot(y,c)
    plt.show()
```

```python
plt.subplot(131) # primer numero indica cuantas filas, segundo indica cuantas columnas
                # ultimo numero indica en que posicion se encontrara el graf
plt.plot(x,y,'r-.')

plt.subplot(132)
plt.plot(x,y,'b-.')

plt.subplot(133)
plt.plot(x,y,'k-.')

plt.show()
```
