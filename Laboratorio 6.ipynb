```python
import scipy.stats
```

```python
# Prueba de hipotesis
# La prueba de levene (de varianzas)
# Ho: varianzas son iguales
# Ha: las varianzas no son iguales


scipy.stats.levene(df[df['Registro de cierre']=='NO']['Avanc financ %'],
                    df[df['Registro de cierre']=='SI']['Avanc financ %'],center='mean')


# el p value es menor a 0.05, por lo que la Ho se rechaza. 
# Significa que las varianzas no son iguales
```

```python
# Prueba T
scipy.stats.ttest_ind(df[df['Registro de cierre']=='NO']['Avanc financ %'],
                        df[df['Registro de cierre']=='SI']['Avanc financ %'],equal_var=False)

# Ho: no hay diferencias significiativas en el promedio de ambos grupos
# Ha: si hay diferencias significas en el promedio de ambos grupos]

# el p value es menor a 0.05 por lo que rechazo la Ho y afirmo que ambos grupos tienen 
# media diferente.
```

```python
# Prueba Anova
#Ho: los promedios de los grupos son iguales
#Ha: al menos uno de los grupos no tiene promedio igual
```

```python
# Crear grupos y variables
df.loc[(df['Avanc financ %']<=40),'rango%']='40 o menos'
df.loc[(df['Avanc financ %']>40) & (df['Avanc financ %']<=50),'rango%']='Entre 40 y 50'
df.loc[(df['Avanc financ %']>50),'rango%']='Mayor a 50'

df.head(10)

```

```python
# Agrupar por rango% a la columna % Var Costo
a=df[df['rango%']=='40 o menos']['Costo actualizado']
b=df[df['rango%']=='Entre 40 y 50']['Costo actualizado']
c=df[df['rango%']=='Mayor a 50']['Costo actualizado']
```

```python
estadistico_f,p_value=scipy.stats.f_oneway(a,b,c)
print ('El estadistico_f es: {0}, el p_value es: {1}'.format(estadistico_f,p_value))

# el valor es menor a 0.05 lo que significa que rechazamos la hipotesis nula
```

```python
# La prueba chi-cuadrado
# Ho: es ambos grupos son independientes (avance financiero y costo actualizado)
# Ha: es que ambos grupos no son independientes
# Crear grupos y variables
df.loc[(df['Avanc financ %']<=50),'rango1%']='50 o menos'
df.loc[(df['Avanc financ %']>50),'rango1%']='Mayor a 50'
df.head()
```

```python
freq_tab=pd.crosstab(df['Registro de cierre'],df['rango1%'])
freq_tab
```

```python
scipy.stats.chi2_contingency(freq_tab,correction=True)

# p value menor a 0.05 se rechaza la Ho

```
